# Task Description
Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, such as the distribution of ages or genders in a population.

## | About the Dataset 
Description - The dataset contains information about the population of various countries over a range of years.
Columns - 
1. Country Name: The name of the country.
2. Country Code: The unique code for each country.
3. Indicator Name: The name of the indicator.
4. Indicator Code: The code for the indicator.
5. Yearly Data (1960-2023): Columns representing the population for each year from 1960 to 2023.

I. Introduction

1. Importing the libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

2. Importing the dataset

```python
df = pd.read_csv("./dataset/world_population.csv")
```

3. Cleaning the dataset

```python
df.isnull().sum()
```

```python
#Unnamed and Empty columns need to be removed
df.drop(columns=["Unnamed: 68","2023"], inplace=True)
df
```

```python
#Removing all the unneeded columns and only keeping the ones that are of use to us
df.drop(columns=["Country Code","Indicator Name","Indicator Code"], inplace=True)
df
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df.info()
```

```python
#As the dataset is very large, we will only consider the population for the first 32 years
df2 = df[['Country Name'] + [str(year) for year in range(1990, 2023)]]
df2
```

```python
#As the dataset contains many rows of data which are not actual countries, we will remove those rows of data as they do not provide any actual insight into the data we require.
#The dataset also contains regions in the world but we are looking to perform the analysis on specific countries therefore, we will filter those out too.
# We define a list of words to filter out
words_to_filter = ['IDA', 'income','World','IBRD','demographic','developed','HIPC','OECD','classification','UN','conflict','and','&','Asia','Africa','Euro','America']
df3 = df2[df2['Country Name'].apply(lambda x: not any(word in x for word in words_to_filter))]
```

II. Data Visualization

    a. Highest Population   

```python
#For analysis, we will see the top 10 countries with the highest population in the year 2022.
final_df = df3.sort_values(by='2022', ascending=False).head(10)
final_df
```

```python
fig, axs = plt.subplots(2, 2, figsize=(14, 10))
axs = axs.flatten()

years = ['1990', '2000', '2010', '2022']

for ax, year in zip(axs, years):
    sns.barplot(data=final_df, x=year, y="Country Name", hue="Country Name", ax=ax, palette="viridis", legend=False)
    ax.set_title(f'Population in {year}', fontsize=14)
    ax.set_xlabel('Population', fontsize=12)
    ax.set_ylabel('Country', fontsize=12)
    ax.tick_params(axis='y', rotation=0) 
    ax.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)
    
plt.tight_layout()
plt.show()
```

```python
years = ['2000', '2010', '2022']

melted_df = final_df.melt(id_vars='Country Name', value_vars=years, var_name='Year', value_name='Population')

sns.set_style("whitegrid")

sns.barplot(data=melted_df, x='Country Name', y='Population', hue='Year', palette='muted')

plt.xticks(rotation=90)

plt.title('Population of Countries in 2000, 2010, and 2022', fontsize=16)
plt.xlabel('Country', fontsize=14)
plt.ylabel('Population', fontsize=14)

plt.legend(title='Year', title_fontsize='14', fontsize='12', loc='upper right')

plt.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)

plt.tight_layout()

plt.show()
```

```python
years = ['2000', '2022']

colors = sns.color_palette("muted", n_colors=len(years))

fig, ax = plt.subplots(figsize=(10, 6))

for year, color in zip(years, colors):
    sns.histplot(final_df[year], bins=30, color=color, label=year, kde=False, alpha=0.6)

plt.title('Distribution of Population Sizes in 2000, 2010, and 2022', fontsize=16)
plt.xlabel('Population', fontsize=14)
plt.ylabel('Number of Countries', fontsize=14)

plt.legend(title='Year', title_fontsize='14', fontsize='12', loc='upper right')

plt.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)

plt.show()
```
